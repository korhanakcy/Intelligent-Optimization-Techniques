**Adƒ±m 4:** (Eƒüer fr ‚â• fn) Daraltma i≈ülemi ùíô ve en iyi ùíôùíè+ùüè ve xr noktalarƒ± arasƒ±nda yapƒ±lmaktadƒ±r. Daraltma (B√ºz√ºlme) i≈ülemini kullanarak C noktasƒ±nƒ±n elde edilmesi i≈üleminde eƒüer R noktasƒ±ndaki fonksiyon deƒüeri W noktasƒ±ndaki fonksiyon deƒüeri ile aynƒ±

**Adƒ±m 5:** √ú√ßgenin B noktasƒ±na doƒüru daraltƒ±lmasƒ±: Eƒüer C noktasƒ±ndaki fonksiyon
deƒüeri W noktasƒ±ndaki fonksiyon deƒüerinden daha k√º√ß√ºk deƒüilse yani f(C)<f(W) ise G ve W noktalarƒ± B noktasƒ±na doƒüru daraltƒ±lmak zorundadƒ±r. Bu daraltma i≈ülemini ilk adƒ±mda G noktasƒ± yerine BG doƒürusunun orta noktasƒ± olan M noktasƒ± ve W noktasƒ± yerine BW doƒürusunun orta noktasƒ± olan S noktasƒ± alƒ±nƒ±r. Sonraki ardƒ±≈üƒ±k i≈ülem adƒ±mlarƒ± B noktasƒ±na doƒüru hep orta noktalar alƒ±narak devam ettirilir. ƒ∞≈ülem B noktasƒ± elde
edildiƒüinde son bulur.

```python
from math import sqrt
from scipy.optimize import minimize, newton

def fitness_function(x):
    x1, x2 = x
    f1 = x1 + 2 * (-x2) + 3
    f2 = 2 * x1 + x2 - 8
    z = f1 ** 2 + f2 ** 2
    return z

# Using Nelder-Mead
initial_guess = [0, 0]
result = minimize(fitness_function, initial_guess, method='Nelder-Mead', options={'maxiter': 1000, 'xatol': 1e-12, 'fatol': 1e-12})
print("Nelder-Mead optimization result:")
print(result)
print("Optimized parameters using Nelder-Mead:")
print(result.x)
print("Minimum value of fitness function using Nelder-Mead:")
print(result.fun)
```
