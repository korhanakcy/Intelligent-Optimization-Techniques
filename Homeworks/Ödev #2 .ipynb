```python
import numpy as np
import matplotlib.pyplot as plt

def calc(c, t, p, p1, p3, k):
    result = c * t * ((p/p1)**((k-1)/k) + (p3/p)**((k-1)/k) - 2)
    return result

# Set the values of the constants
c = 1
t = 1
p1 = 2
p3 = 3
k = 10

# Define the range of values for p
p_values = np.linspace(0.1, 10, 100)

# Calculate the function values for each value of p
function_values = [calc(c, t, p, p1, p3, k) for p in p_values]

# Plot the function values against p
plt.plot(p_values, function_values)
plt.xlabel('p')
plt.ylabel('Function value')
plt.title('Plot of function against p')
plt.show()
```
