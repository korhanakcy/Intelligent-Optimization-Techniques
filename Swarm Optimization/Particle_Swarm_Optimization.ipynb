# **Par√ßacƒ±k S√ºr√º Optimizasyonu(Particle Swarm Optimization)**

Par√ßacƒ±k s√ºr√º optimizasyonu, balƒ±k ve ku≈ü gibi canlƒ±larƒ±n doƒüadaki sosyal davranƒ±≈ülarƒ±ndan esinlenerek geli≈ütirilen meta-sezgisel bir optimizasyon algoritmasƒ±dƒ±r(Eberhart ve Kennedy, 1995). Par√ßacƒ±k s√ºr√º optimizasyonunun k√∂klerini
iki ana bile≈üene dayandƒ±rƒ±r. Bunlardan birisi yapay ya≈üam √∂r√ºnt√ºleri, diƒüeri ise evrimsel
hesaplamadƒ±r. Bu y√ºzden, par√ßacƒ±k s√ºr√º optimizasyonu, hem genetik algoritmalarla hem
de evrimsel programlama ile doƒürudan ili≈ükilidir.

 Bir ku≈ü s√ºr√ºs√º birbirine √ßarpmadan birlikte u√ßar, y√∂n deƒüi≈ütirir ve bunlarƒ± e≈üzamanlƒ± olarak ger√ßekle≈ütirir. S√ºr√ºdeki bir ku≈ü yiyecek bulduƒüunda, diƒüer ku≈ülarla ileti≈üim kurarak onlarƒ±n da yiyeceƒüe ula≈ümasƒ±nƒ± saƒülar. Aynƒ± ≈üekilde, balƒ±k s√ºr√ºlerinde de bir balƒ±k bir tehlike sezdiƒüinde s√ºr√ºdeki diƒüer balƒ±klarla ileti≈üime ge√ßerek onlarƒ±n tehlikeden haberdar olmasƒ±nƒ± saƒülar. √ñzellikle ku≈ü ve balƒ±k s√ºr√ºlerinin i≈ü birlik√ßi davranƒ±≈ülarƒ±ndan ilham alan par√ßacƒ±k s√ºr√º optimizasyonu, yapay par√ßacƒ±klarƒ±n arama uzayƒ±nda ku≈ü ve balƒ±k s√ºr√ºlerinin hƒ±z ve konum gibi fiziksel √∂zelliklerini taklit etmesiyle optimal √ß√∂z√ºm√º bulmayƒ± ama√ßlar. Algoritmada, her par√ßacƒ±k muhtemel bir √ß√∂z√ºm√º temsil etmektedir (Eberhart ve Kennedy, 1995; Kaewkamnerdpong ve Bentley, 2005; Krause vd.,
2013; Vasuki, 2020). 

Bir s√ºr√º modeli geli≈ütirmek i√ßin belirli prensiplerin saƒülanmasƒ± gerekmektedir. Bunlar,

‚Ä¢ √áoklu birimler arasƒ±nda bilgi payla≈üƒ±mƒ±,

‚Ä¢ Birimlerin kendi kendini organize etmesi ve evrim ge√ßirmesi,

‚Ä¢ Birimlerin e≈ü √∂ƒürenmesi a√ßƒ±sƒ±ndan etkili olmasƒ±,

‚Ä¢ Uygulamalƒ± ve ger√ßek zamanlƒ± problemlere kolayca uyarlanabilmesi

Par√ßacƒ±klarƒ±n arama uzayƒ±ndaki konumlarƒ±nƒ±n deƒüi≈üimi bireylerin sosyo-bili≈üsel eƒüilimine baƒülƒ±dƒ±r ve par√ßacƒ±klarƒ±n mevcut konumlarƒ±nƒ±n g√ºncellenmesi i√ßin hƒ±z vekt√∂r√º kullanƒ±lƒ±r. Bir par√ßacƒ±ƒüƒ±n en iyi konumu ‚Äúpbest‚Äù olarak adlandƒ±rƒ±lƒ±r ve bu deƒüer hƒ±z
vekt√∂r√ºn√ºn g√ºncellenmesi i√ßin hafƒ±zada tutulur. Bununla birlikte, hƒ±z vekt√∂r√ºn√ºn hesaplanmasƒ±nda pop√ºlasyondaki herhangi bir par√ßacƒ±k tarafƒ±ndan elde edilen en iyi konum da kullanƒ±lmaktadƒ±r. Bu konum ise ‚Äúgbest‚Äù olarak adlandƒ±rƒ±lmaktadƒ±r. ‚Äúpbest‚Äù ve ‚Äúgbest‚Äù
deƒüerleri bulunduktan sonra par√ßacƒ±klarƒ±n konumu ve hƒ±zƒ± g√ºncellenir (Eberhart ve Kennedy,

# **√ñrnek 1**

f(x,y)=(x-2y+3)^2+(2x+y-8)^2 fonksiyonunun. Global minimum noktasƒ±nƒ±n bulunmasƒ± √ºzerine bir √∂rnektir. Bu √∂rnekte, t√ºm par√ßacƒ±klar rastgele noktalardan, deƒüerin 0'a yakƒ±n olduƒüu x ve y'nin optimal konumuna hareket etmelidir.

par√ßacƒ±k sayƒ±sƒ± : i

Boyut sayƒ±sƒ± : n

Uygunluk fonksiyonu : f(x_i)

Par√ßacƒ±klar : x_i = (x_i1, x_i2, ..., x_in)

Mevcut hƒ±z : v_i = (v_i1, v_i2, ..., v_in)

Bireysel par√ßacƒ±ƒüƒ±n en iyisi : p_i = (p_i1, p_i2, ..., p_in)

K√ºresel par√ßacƒ±klarƒ±n en iyisi: p_g = (p_g1, p_g2, ..., p_gn)

Eylemsizlik bile≈üeni : w * v_i(t)

Bili≈üsel bile≈üen : c_1 * r_1 * (p_i - x_i(t))

Sosyal bile≈üen : c_2 * r_2 * (g_i - x_i(t))

Hƒ±z ayarƒ± : v_i(t+1) <- Inertia+Cognitive+Social

Konum ayarƒ± : x_i(t+1) <- x_i(t)+v_i(t+1)

**Algoritma Akƒ±≈ü Diyagramƒ±:**

1. x_i par√ßacƒ±k pop√ºlasyon dizisini ba≈ülatƒ±n
2. D√∂ng√º
3. Her par√ßacƒ±k i√ßin uygunluk fonksiyonunu f(x_i) kullanarak uygunluƒüu hesaplayƒ±n.
4. Mevcut uygunluk deƒüerini en iyi p_i deƒüeriyle kar≈üƒ±la≈ütƒ±rƒ±n. En iyiden daha iyiyse, en iyiyi ge√ßerli x_i deƒüeriyle deƒüi≈ütirin.
5. S√ºr√ºn√ºn en iyi par√ßacƒ±ƒüƒ±nƒ± bireysel par√ßacƒ±ƒüƒ±n en iyisinden kontrol edin ve en iyi diziyi global en iyi p_g'ye atayƒ±n.
6. v_i(t+1) hƒ±zƒ±nƒ± hesaplayƒ±n ve par√ßacƒ±klarƒ±n konumunu x_i(t+1) olarak g√ºncelleyin
7. Bir kriter kar≈üƒ±lanƒ±rsa d√∂ng√ºden √ßƒ±kƒ±n.
8. D√∂ng√ºy√º sonlandƒ±r

```python
import random
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# Uygunluk Fonksiyonu
# Problemin a≈üaƒüƒ±daki denklemle ifade edilebileceƒüini d√º≈ü√ºnelim;
# f(x1,x2)=(x1+2*-x2+3)^2 + (2*x1+x2-8)^2
# Ama√ß, 0 olan bir minimum bulmaktƒ±r.

def fitness_function(x1,x2):
  f1=x1+2*-x2+3
  f2=2*x1+x2-8
  z = f1**2+f2**2
  return z

def update_velocity(particle, velocity, pbest, gbest, w_min=0.5, max=1.0, c=0.1):

  # Yeni hƒ±z dizisini ba≈ülatƒ±lmasƒ±
  
  num_particle = len(particle)
  new_velocity = np.array([0.0 for i in range(num_particle)])
  
  # Normal daƒüƒ±lƒ±mdan rastgele r1, r2 ve inertia aƒüƒ±rlƒ±ƒüƒ± √ºretilmesi
  
  r1 = random.uniform(0,max)
  r2 = random.uniform(0,max)
  w = random.uniform(w_min,max)
  c1 = c
  c2 = c

  # Yeni hƒ±zƒ±n hesapla
  
  for i in range(num_particle):
    new_velocity[i] = w*velocity[i] + c1*r1*(pbest[i]-particle[i])+c2*r2*(gbest[i]-particle[i])
  return new_velocity

def update_position(particle, velocity):

  # Par√ßacƒ±klarƒ± hƒ±z ekleyerek hareket ettirilmesi
  
  new_particle = particle + velocity
  return new_particle


def pso_2d(population, dimension, position_min, position_max, generation, fitness_criterion):
  # Ba≈ülatƒ±lmasƒ±
  # Pop√ºlasyon
  particles = [[random.uniform(position_min, position_max) for j in range(dimension)] for i in range(population)]
  # Par√ßacƒ±klarƒ±n en iyi konumlarƒ±
  pbest_position = particles
  # Uygunluk
  pbest_fitness = [fitness_function(p[0],p[1]) for p in particles]
  # En iyi par√ßacƒ±ƒüƒ±n indeksi
  gbest_index = np.argmin(pbest_fitness)
  # Global en iyi par√ßacƒ±k konumu
  gbest_position = pbest_position[gbest_index]
  # Hƒ±z
  velocity = [[0.0 for j in range(dimension)] for i in range(population)]
  
  # Jenerasyon sayƒ±sƒ± i√ßin d√∂ng√º
  for t in range(generation):
    # Ortalama uygunluk deƒüeri √∂nceden tanƒ±mlanmƒ±≈ü bir ba≈üarƒ± kriterine ula≈üƒ±rsa durdur
    if np.average(pbest_fitness) <= fitness_criterion:
          break
    else:
      for n in range(population):
        # Her par√ßacƒ±ƒüƒ±n hƒ±zƒ±nƒ± g√ºncellenmesi
        velocity[n] = update_velocity(particles[n], velocity[n], pbest_position[n], gbest_position)
        # Par√ßacƒ±klarƒ± yeni konuma ta≈üƒ±nmasƒ±
        particles[n] = update_position(particles[n], velocity[n])
    # Uygunluk deƒüerini hesaplanmasƒ±
    pbest_fitness = [fitness_function(p[0],p[1]) for p in particles]
    # En iyi par√ßacƒ±ƒüƒ±n indeksini bulunmasƒ±
    gbest_index = np.argmin(pbest_fitness)
    # En iyi par√ßacƒ±ƒüƒ±n konumunu g√ºncellenmesi
    gbest_position = pbest_position[gbest_index]

  print('Global En ƒ∞yi Konum: ', gbest_position)
  print('En Uygun Deƒüer: ', min(pbest_fitness))
  print('Ortalama Par√ßacƒ±k i√ßin En ƒ∞yi Uygunluk Deƒüeri: ', np.average(pbest_fitness))
  print('Jenerasyon Sayƒ±sƒ±: ', t)

pso_2d(population = 100, dimension = 2, position_min = -100.0, position_max = 100.0, generation = 400 ,fitness_criterion = 10e-4)
```

```python
from __future__ import print_function    

import time
import math
import numpy as np
import pylab as py

# Control parameters
w = 0.5                   # Ara aƒüƒ±rlƒ±k. Bazƒ± varyasyonlarda, iterasyon sayƒ±sƒ±na g√∂re deƒüi≈üecek ≈üekilde ayarlanmƒ±≈ütƒ±r.
c1 = 2.0                  # Bir par√ßacƒ±ƒüƒ±n bulduƒüu optimuma dayalƒ± aramanƒ±n aƒüƒ±rlƒ±ƒüƒ±
c2 = 2.0                  # S√ºr√º tarafƒ±ndan bulunan optimuma dayalƒ± olarak aramanƒ±n aƒüƒ±rlƒ±ƒüƒ±
v_fct = 1                 # Hƒ±z ayar fakt√∂r√º. Standart PSO i√ßin 1'e ayarlanmalƒ±


Np = 40                   # populasyon (par√ßacƒ±k sayƒ±sƒ±)
D = 4                     # boyut (uygunluk fonksiyonundaki parametre sayƒ±sƒ±)
max_iter = 100            # maksimum iterasyon sayƒ±sƒ±
xL = np.zeros(D) - 4      # alt sƒ±nƒ±r (homojen olmasƒ± gerekmez)
xU = np.zeros(D) + 4      # √ºst sƒ±nƒ±r (homojen olmasƒ± gerekmez)   



# Fitness fonksiyonu. Kod, uygunluk i≈ülevinin deƒüerini en √ºst d√ºzeye √ßƒ±karƒ±r
def fitness(x):
    # x, D x Np boyutunda bir matristir. T√ºm s√ºr√ºn√ºn konumu bir kerede girilir. 
    # B√∂ylece, bir i≈ülev √ßaƒürƒ±sƒ± t√ºm s√ºr√ºn√ºn uygunluk deƒüerini deƒüerlendirir.
    # F, Np boyutunda bir vekt√∂rd√ºr. Her eleman, s√ºr√ºdeki her par√ßacƒ±ƒüƒ±n uygunluk deƒüerini temsil eder.
    
    F_sphere = 2.0 - np.sum(np.multiply(x,x),0)    # modified sphere function
    return F_sphere



pbest_val = np.zeros(Np)            # Ki≈üisel en iyi uygunluk deƒüeri. Par√ßacƒ±k ba≈üƒ±na bir pbest deƒüeri.
gbest_val = np.zeros(max_iter)      # Global en iyi uygunluk deƒüeri. Yineleme ba≈üƒ±na bir gbest deƒüeri.


pbest = np.zeros((D,Np))            # pbest √ß√∂z√ºm√º
gbest = np.zeros(D)                 # gbest √ß√∂z√ºm√º

gbest_store = np.zeros((D,max_iter))   # her yinelemede gbest √ß√∂z√ºm√ºn√ºn depolanmasƒ±

pbest_val_avg_store = np.zeros(max_iter)
fitness_avg_store = np.zeros(max_iter)

x = np.random.rand(D,Np)            # Par√ßacƒ±klarƒ±n ba≈ülangƒ±√ß konumu
v = np.zeros((D,Np))                # Par√ßacƒ±klarƒ±n ba≈ülangƒ±√ß hƒ±zƒ±


# Par√ßacƒ±klarƒ±n ba≈ülangƒ±√ß konumunun verilen sƒ±nƒ±rlar √ºzerinde ayarlanmasƒ± [xL,xU]
for m in range(D):    
    x[m,:] = xL[m] + (xU[m]-xL[m])*x[m,:]
    

# ƒ∞lk deƒüerlendirmeler (iterasyon = 0 i√ßin)
#  ƒ∞lk s√ºr√ºlerin uygunluƒüunu deƒüerlendirilmesi   
fit = fitness(x)           # Np boyutunda vekt√∂r

pbest_val = np.copy(fit)   # ilk ki≈üisel en iyi = ba≈ülangƒ±√ß uygunluk deƒüerleri. Np boyutunda vekt√∂r
pbest = np.copy(x)         #ilk pbest √ß√∂z√ºm√º = ba≈ülangƒ±√ß konumu. D x Np boyut matris

# gbest_val ve gbest hesaplanƒ±yor. gbest'in pbest i√ßindeki en iyi √ß√∂z√ºmd√ºr.                                                                                                                      
ind = np.argmax(pbest_val)                # pbest_val'in maksimum olduƒüu dizin. 
gbest_val[0] = np.copy(pbest_val[ind])    # ilk gbest_val deƒüerini ayarlanmasƒ±
gbest = np.copy(pbest[:,ind])
pbest_val_avg_store[0] = np.mean(pbest_val)
fitness_avg_store[0] = np.mean(fit)

print("Iter. =", 0, ". gbest_val = ", gbest_val[0])
print("gbest_val = ",gbest_val[0])


for iter in range(1,max_iter):
    
  
    r1 = np.random.rand(D,Np)           
    r2 = np.random.rand(D,Np)              
    v_global = np.multiply(((x.transpose()-gbest).transpose()),r2)*c2*(-1.0)    # global optimuma doƒüru hƒ±z    
    v_local = np.multiply((pbest- x),r1)*c1           # yerel optimuma doƒüru hƒ±z (pbest)

    v = w*v + (v_local + v_global)      # hƒ±z g√ºncellemesi
  
    x = x + v*v_fct                     # pozisyon g√ºncellemesi
    
    
    fit = fitness(x)                    # fitness fonksiyonunun √ßaƒürƒ±lmasƒ±.
    
    # pbest ve pbest_val deƒüelerinin g√ºncellenmesi
    ind = np.argwhere(fit > pbest_val)  # mevcut uygunluk deƒüeri setinin pbset deƒüerinden b√ºy√ºk olduƒüu indeksler
    pbest_val[ind] = np.copy(fit[ind])  # fit > pbest_val olan par√ßacƒ±k indekslerinde pbset_val'i g√ºncellenmesi
    pbest[:,ind] = np.copy(x[:,ind])    # fiit > pbest_val olan par√ßacƒ±k indeksleri i√ßin pbest'i g√ºncellenmesi
  
    
    # gbest and gbest_val update
    ind2 = np.argmax(pbest_val)                       # fitnessin maksimum olduƒüu dizin
    gbest_val[iter] = np.copy(pbest_val[ind2])        # her yinelemede gbest deƒüeri
    gbest = np.copy(pbest[:,ind2])                    # en iyi global √ß√∂z√ºm -gbest-
    
    gbest_store[:,iter] = np.copy(gbest)              # gbest √ß√∂z√ºm√ºn√ºn depolanmasƒ±
    
    pbest_val_avg_store[iter] = np.mean(pbest_val)
    fitness_avg_store[iter] = np.mean(fit)
    print("Iter. =", iter, ". gbest_val = ", gbest_val[iter])  

# Plotting
py.close('all')
py.figure(1)
py.plot(gbest_val,label = 'gbest_val')
py.plot(pbest_val_avg_store, label = 'Avg. pbest')
py.plot(fitness_avg_store, label = 'Avg. fitness')
py.legend()


py.xlabel('iterations')
py.ylabel('fitness, gbest_val')


py.figure(2)
for m in range(D):
    py.plot(gbest_store[m,:],label = 'D = ' + str(m+1))
    
py.legend()
py.xlabel('iterations')
py.ylabel('Best solution, gbest[:,iter]')
```
