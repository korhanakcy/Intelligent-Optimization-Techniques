# **Parçacık Sürü Optimizasyonu(Particle Swarm Optimization)**

Parçacık sürü optimizasyonu, balık ve kuş gibi canlıların doğadaki sosyal davranışlarından esinlenerek geliştirilen meta-sezgisel bir optimizasyon algoritmasıdır(Eberhart ve Kennedy, 1995). Parçacık sürü optimizasyonunun köklerini
iki ana bileşene dayandırır. Bunlardan birisi yapay yaşam örüntüleri, diğeri ise evrimsel
hesaplamadır. Bu yüzden, parçacık sürü optimizasyonu, hem genetik algoritmalarla hem
de evrimsel programlama ile doğrudan ilişkilidir.

 Bir kuş sürüsü birbirine çarpmadan birlikte uçar, yön değiştirir ve bunları eşzamanlı olarak gerçekleştirir. Sürüdeki bir kuş yiyecek bulduğunda, diğer kuşlarla iletişim kurarak onların da yiyeceğe ulaşmasını sağlar. Aynı şekilde, balık sürülerinde de bir balık bir tehlike sezdiğinde sürüdeki diğer balıklarla iletişime geçerek onların tehlikeden haberdar olmasını sağlar. Özellikle kuş ve balık sürülerinin iş birlikçi davranışlarından ilham alan parçacık sürü optimizasyonu, yapay parçacıkların arama uzayında kuş ve balık sürülerinin hız ve konum gibi fiziksel özelliklerini taklit etmesiyle optimal çözümü bulmayı amaçlar. Algoritmada, her parçacık muhtemel bir çözümü temsil etmektedir (Eberhart ve Kennedy, 1995; Kaewkamnerdpong ve Bentley, 2005; Krause vd.,
2013; Vasuki, 2020). 

Bir sürü modeli geliştirmek için belirli prensiplerin sağlanması gerekmektedir. Bunlar,

• Çoklu birimler arasında bilgi paylaşımı,

• Birimlerin kendi kendini organize etmesi ve evrim geçirmesi,

• Birimlerin eş öğrenmesi açısından etkili olması,

• Uygulamalı ve gerçek zamanlı problemlere kolayca uyarlanabilmesi

Parçacıkların arama uzayındaki konumlarının değişimi bireylerin sosyo-bilişsel eğilimine bağlıdır ve parçacıkların mevcut konumlarının güncellenmesi için hız vektörü kullanılır. Bir parçacığın en iyi konumu “pbest” olarak adlandırılır ve bu değer hız
vektörünün güncellenmesi için hafızada tutulur. Bununla birlikte, hız vektörünün hesaplanmasında popülasyondaki herhangi bir parçacık tarafından elde edilen en iyi konum da kullanılmaktadır. Bu konum ise “gbest” olarak adlandırılmaktadır. “pbest” ve “gbest”
değerleri bulunduktan sonra parçacıkların konumu ve hızı güncellenir (Eberhart ve Kennedy,

# **Örnek 1**

f(x,y)=(x-2y+3)^2+(2x+y-8)^2 fonksiyonunun. Global minimum noktasının bulunması üzerine bir örnektir. Bu örnekte, tüm parçacıklar rastgele noktalardan, değerin 0'a yakın olduğu x ve y'nin optimal konumuna hareket etmelidir.

parçacık sayısı : i

Boyut sayısı : n

Uygunluk fonksiyonu : f(x_i)

Parçacıklar : x_i = (x_i1, x_i2, ..., x_in)

Mevcut hız : v_i = (v_i1, v_i2, ..., v_in)

Bireysel parçacığın en iyisi : p_i = (p_i1, p_i2, ..., p_in)

Küresel parçacıkların en iyisi: p_g = (p_g1, p_g2, ..., p_gn)

Eylemsizlik bileşeni : w * v_i(t)

Bilişsel bileşen : c_1 * r_1 * (p_i - x_i(t))

Sosyal bileşen : c_2 * r_2 * (g_i - x_i(t))

Hız ayarı : v_i(t+1) <- Inertia+Cognitive+Social

Konum ayarı : x_i(t+1) <- x_i(t)+v_i(t+1)

**Algoritma Akış Diyagramı:**

1. x_i parçacık popülasyon dizisini başlatın
2. Döngü
3. Her parçacık için uygunluk fonksiyonunu f(x_i) kullanarak uygunluğu hesaplayın.
4. Mevcut uygunluk değerini en iyi p_i değeriyle karşılaştırın. En iyiden daha iyiyse, en iyiyi geçerli x_i değeriyle değiştirin.
5. Sürünün en iyi parçacığını bireysel parçacığın en iyisinden kontrol edin ve en iyi diziyi global en iyi p_g'ye atayın.
6. v_i(t+1) hızını hesaplayın ve parçacıkların konumunu x_i(t+1) olarak güncelleyin
7. Bir kriter karşılanırsa döngüden çıkın.
8. Döngüyü sonlandır

```python
import random
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# Uygunluk Fonksiyonu
# Problemin aşağıdaki denklemle ifade edilebileceğini düşünelim;
# f(x1,x2)=(x1+2*-x2+3)^2 + (2*x1+x2-8)^2
# Amaç, 0 olan bir minimum bulmaktır.

def fitness_function(x1,x2):
  f1=x1+2*-x2+3
  f2=2*x1+x2-8
  z = f1**2+f2**2
  return z

def update_velocity(particle, velocity, pbest, gbest, w_min=0.5, max=1.0, c=0.1):

  # Yeni hız dizisini başlatılması
  
  num_particle = len(particle)
  new_velocity = np.array([0.0 for i in range(num_particle)])
  
  # Normal dağılımdan rastgele r1, r2 ve inertia ağırlığı üretilmesi
  
  r1 = random.uniform(0,max)
  r2 = random.uniform(0,max)
  w = random.uniform(w_min,max)
  c1 = c
  c2 = c

  # Yeni hızın hesapla
  
  for i in range(num_particle):
    new_velocity[i] = w*velocity[i] + c1*r1*(pbest[i]-particle[i])+c2*r2*(gbest[i]-particle[i])
  return new_velocity

def update_position(particle, velocity):

  # Parçacıkları hız ekleyerek hareket ettirilmesi
  
  new_particle = particle + velocity
  return new_particle


def pso_2d(population, dimension, position_min, position_max, generation, fitness_criterion):
  # Başlatılması
  # Popülasyon
  particles = [[random.uniform(position_min, position_max) for j in range(dimension)] for i in range(population)]
  # Parçacıkların en iyi konumları
  pbest_position = particles
  # Uygunluk
  pbest_fitness = [fitness_function(p[0],p[1]) for p in particles]
  # En iyi parçacığın indeksi
  gbest_index = np.argmin(pbest_fitness)
  # Global en iyi parçacık konumu
  gbest_position = pbest_position[gbest_index]
  # Hız
  velocity = [[0.0 for j in range(dimension)] for i in range(population)]
  
  # Jenerasyon sayısı için döngü
  for t in range(generation):
    # Ortalama uygunluk değeri önceden tanımlanmış bir başarı kriterine ulaşırsa durdur
    if np.average(pbest_fitness) <= fitness_criterion:
          break
    else:
      for n in range(population):
        # Her parçacığın hızını güncellenmesi
        velocity[n] = update_velocity(particles[n], velocity[n], pbest_position[n], gbest_position)
        # Parçacıkları yeni konuma taşınması
        particles[n] = update_position(particles[n], velocity[n])
    # Uygunluk değerini hesaplanması
    pbest_fitness = [fitness_function(p[0],p[1]) for p in particles]
    # En iyi parçacığın indeksini bulunması
    gbest_index = np.argmin(pbest_fitness)
    # En iyi parçacığın konumunu güncellenmesi
    gbest_position = pbest_position[gbest_index]

  print('Global En İyi Konum: ', gbest_position)
  print('En Uygun Değer: ', min(pbest_fitness))
  print('Ortalama Parçacık için En İyi Uygunluk Değeri: ', np.average(pbest_fitness))
  print('Jenerasyon Sayısı: ', t)

pso_2d(population = 100, dimension = 2, position_min = -100.0, position_max = 100.0, generation = 400 ,fitness_criterion = 10e-4)
```

# **Örnek 2**

```python
from __future__ import print_function    

import time
import math
import numpy as np
import pylab as py

# Control parameters
w = 0.5                   # Ara ağırlık. Bazı varyasyonlarda, iterasyon sayısına göre değişecek şekilde ayarlanmıştır.
c1 = 2.0                  # Bir parçacığın bulduğu optimuma dayalı aramanın ağırlığı
c2 = 2.0                  # Sürü tarafından bulunan optimuma dayalı olarak aramanın ağırlığı
v_fct = 1                 # Hız ayar faktörü. Standart PSO için 1'e ayarlanmalı


Np = 40                   # populasyon (parçacık sayısı)
D = 4                     # boyut (uygunluk fonksiyonundaki parametre sayısı)
max_iter = 100            # maksimum iterasyon sayısı
xL = np.zeros(D) - 4      # alt sınır (homojen olması gerekmez)
xU = np.zeros(D) + 4      # üst sınır (homojen olması gerekmez)   



# Fitness fonksiyonu. Kod, uygunluk işlevinin değerini en üst düzeye çıkarır
def fitness(x):
    # x, D x Np boyutunda bir matristir. Tüm sürünün konumu bir kerede girilir. 
    # Böylece, bir işlev çağrısı tüm sürünün uygunluk değerini değerlendirir.
    # F, Np boyutunda bir vektördür. Her eleman, sürüdeki her parçacığın uygunluk değerini temsil eder.
    
    F_sphere = 2.0 - np.sum(np.multiply(x,x),0)    # modified sphere function
    return F_sphere



pbest_val = np.zeros(Np)            # Kişisel en iyi uygunluk değeri. Parçacık başına bir pbest değeri.
gbest_val = np.zeros(max_iter)      # Global en iyi uygunluk değeri. Yineleme başına bir gbest değeri.


pbest = np.zeros((D,Np))            # pbest çözümü
gbest = np.zeros(D)                 # gbest çözümü

gbest_store = np.zeros((D,max_iter))   # her yinelemede gbest çözümünün depolanması

pbest_val_avg_store = np.zeros(max_iter)
fitness_avg_store = np.zeros(max_iter)

x = np.random.rand(D,Np)            # Parçacıkların başlangıç konumu
v = np.zeros((D,Np))                # Parçacıkların başlangıç hızı


# Parçacıkların başlangıç konumunun verilen sınırlar üzerinde ayarlanması [xL,xU]
for m in range(D):    
    x[m,:] = xL[m] + (xU[m]-xL[m])*x[m,:]
    

# İlk değerlendirmeler (iterasyon = 0 için)
#  İlk sürülerin uygunluğunu değerlendirilmesi   
fit = fitness(x)           # Np boyutunda vektör

pbest_val = np.copy(fit)   # ilk kişisel en iyi = başlangıç uygunluk değerleri. Np boyutunda vektör
pbest = np.copy(x)         #ilk pbest çözümü = başlangıç konumu. D x Np boyut matris

# gbest_val ve gbest hesaplanıyor. gbest'in pbest içindeki en iyi çözümdür.                                                                                                                      
ind = np.argmax(pbest_val)                # pbest_val'in maksimum olduğu dizin. 
gbest_val[0] = np.copy(pbest_val[ind])    # ilk gbest_val değerini ayarlanması
gbest = np.copy(pbest[:,ind])
pbest_val_avg_store[0] = np.mean(pbest_val)
fitness_avg_store[0] = np.mean(fit)

print("Iter. =", 0, ". gbest_val = ", gbest_val[0])
print("gbest_val = ",gbest_val[0])


for iter in range(1,max_iter):
    
  
    r1 = np.random.rand(D,Np)           
    r2 = np.random.rand(D,Np)              
    v_global = np.multiply(((x.transpose()-gbest).transpose()),r2)*c2*(-1.0)    # global optimuma doğru hız    
    v_local = np.multiply((pbest- x),r1)*c1           # yerel optimuma doğru hız (pbest)

    v = w*v + (v_local + v_global)      # hız güncellemesi
  
    x = x + v*v_fct                     # pozisyon güncellemesi
    
    
    fit = fitness(x)                    # fitness fonksiyonunun çağrılması.
    
    # pbest ve pbest_val değelerinin güncellenmesi
    ind = np.argwhere(fit > pbest_val)  # mevcut uygunluk değeri setinin pbset değerinden büyük olduğu indeksler
    pbest_val[ind] = np.copy(fit[ind])  # fit > pbest_val olan parçacık indekslerinde pbset_val'i güncellenmesi
    pbest[:,ind] = np.copy(x[:,ind])    # fiit > pbest_val olan parçacık indeksleri için pbest'i güncellenmesi
  
    
    # gbest and gbest_val update
    ind2 = np.argmax(pbest_val)                       # fitnessin maksimum olduğu dizin
    gbest_val[iter] = np.copy(pbest_val[ind2])        # her yinelemede gbest değeri
    gbest = np.copy(pbest[:,ind2])                    # en iyi global çözüm -gbest-
    
    gbest_store[:,iter] = np.copy(gbest)              # gbest çözümünün depolanması
    
    pbest_val_avg_store[iter] = np.mean(pbest_val)
    fitness_avg_store[iter] = np.mean(fit)
    print("Iter. =", iter, ". gbest_val = ", gbest_val[iter])  

# Plotting
py.close('all')
py.figure(1)
py.plot(gbest_val,label = 'gbest_val')
py.plot(pbest_val_avg_store, label = 'Avg. pbest')
py.plot(fitness_avg_store, label = 'Avg. fitness')
py.legend()


py.xlabel('iterations')
py.ylabel('fitness, gbest_val')


py.figure(2)
for m in range(D):
    py.plot(gbest_store[m,:],label = 'D = ' + str(m+1))
    
py.legend()
py.xlabel('iterations')
py.ylabel('Best solution, gbest[:,iter]')
```
